```python
import pandas as pd
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Télécharger les ressources nécessaires de NLTK
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Définir les stopwords en anglais
stop_words = set(stopwords.words('english'))

# Initialiser le lemmatizer
lemmatizer = WordNetLemmatizer()

# Lire le fichier CSV avec le bon encodage
df = pd.read_csv('spam.csv', encoding='latin1')

# Nettoyer et lemmatizer la colonne 'v1'
cleaned_v1 = []
for text in df['v1']:
    # Vérifier si le texte est non nul
    if pd.isna(text):
        cleaned_v1.append("")
        continue
    
    # Supprimer les caractères spéciaux et les chiffres
    text = re.sub(r'[^A-Za-z\s]', '', text)
    
    # Convertir le texte en minuscules
    text = text.lower()
    
    # Tokenizer le texte
    words = nltk.word_tokenize(text)
    
    # Éliminer les stopwords et lemmatizer les mots restants
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    
    # Rejoindre les mots en une seule chaîne
    cleaned_text = ' '.join(lemmatized_words)
    
    cleaned_v1.append(cleaned_text)

# Nettoyer et lemmatizer la colonne 'v2'
cleaned_v2 = []
for text in df['v2']:
    # Vérifier si le texte est non nul
    if pd.isna(text):
        cleaned_v2.append("")
        continue
    
    # Supprimer les caractères spéciaux et les chiffres
    text = re.sub(r'[^A-Za-z\s]', '', text)
    
    # Convertir le texte en minuscules
    text = text.lower()
    
    # Tokenizer le texte
    words = nltk.word_tokenize(text)
    
    # Éliminer les stopwords et lemmatizer les mots restants
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    
    # Rejoindre les mots en une seule chaîne
    cleaned_text = ' '.join(lemmatized_words)
    
    cleaned_v2.append(cleaned_text)

# Ajouter les nouvelles colonnes au dataframe
df['cleaned_v1'] = cleaned_v1
df['cleaned_v2'] = cleaned_v2

# Afficher les résultats
print(df[['v1', 'cleaned_v1', 'v2', 'cleaned_v2']].head())
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Lire le fichier CSV avec les données nettoyées
df = pd.read_csv('spam.csv', encoding='latin1')

# Concaténer les colonnes nettoyées en une seule chaîne de texte
all_text = ' '.join(df['cleaned_v1'].tolist() + df['cleaned_v2'].tolist())

# Spécifier le chemin de la police TrueType (.ttf)
font_path = 'DejaVuSans.ttf'  # Remplacez par le chemin de votre police TrueType

# Générer le nuage de mots en utilisant la police TrueType spécifiée
wordcloud = WordCloud(width=800, height=400, background_color='white', font_path=font_path).generate(all_text)

# Afficher le nuage de mots
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Définir les paramètres de CountVectorizer pour les n-grammes
ngram_range = (1, )  # Utilisation de n-grammes de 1 à 2 mots
max_features = 2000  # Limite le nombre de features pour éviter les données trop denses

# Initialiser CountVectorizer
vectorizer = CountVectorizer(ngram_range=ngram_range, max_features=max_features)

# Appliquer CountVectorizer aux données nettoyées
X = vectorizer.fit_transform(df['cleaned_v1'])

# Obtenir les noms des features (mots) extraits
feature_names = vectorizer.get_feature_names_out()

# Créer un DataFrame pour les n-grammes avec leurs fréquences
ngram_df = pd.DataFrame(X.toarray(), columns=feature_names)

# Afficher les premières lignes du DataFrame des n-grammes
print(ngram_df.head())
```

```python
import matplotlib.pyplot as plt

# Calculer la longueur des messages originaux et nettoyés
df['original_length'] = df['v1'].apply(len)
df['cleaned_length'] = df['cleaned_v1'].apply(len)

# Créer un histogramme pour comparer les longueurs
plt.figure(figsize=(10, 6))
plt.hist(df['original_length'], bins=50, alpha=0.5, label='Original Messages')
plt.hist(df['cleaned_length'], bins=50, alpha=0.5, label='Cleaned Messages')
plt.title('Distribution des Longueurs de Messages')
plt.xlabel('Longueur des Messages')
plt.ylabel('Fréquence')
plt.legend()
plt.show()
```

```python
# Compter les occurrences de chaque type de message
message_types = df['v1'].value_counts()

# Tracer un diagramme à barres
plt.figure(figsize=(8, 6))
message_types.plot(kind='bar', color=['blue', 'green'])
plt.title('Répartition des Types de Messages')
plt.xlabel('Type de Message')
plt.ylabel('Nombre de Messages')
plt.xticks(rotation=0)
plt.show()
```

```python
from collections import Counter

# Tokeniser tous les messages nettoyés
all_words = ' '.join(df['cleaned_v1']).split()

# Compter la fréquence des mots
word_freq = Counter(all_words)

# Obtenir les mots les plus fréquents
common_words = word_freq.most_common(20)

print(common_words)

# Créer un diagramme à barres pour les mots les plus fréquents
plt.figure(figsize=(10, 6))
plt.bar([word[0] for word in common_words], [count[1] for count in common_words], color='skyblue')
plt.title('Mots les Plus Fréquents dans les Messages')
plt.xlabel('Mots')
plt.ylabel('Fréquence')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# Télécharger les ressources nécessaires de NLTK
nltk.download('stopwords')
nltk.download('punkt')

# Définir les stopwords en anglais
stop_words = set(stopwords.words('english'))

# Initialiser le stemmer
stemmer = PorterStemmer()

# Nettoyer et stemmer la colonne 'v1'
cleaned_v1_stem = []
for text in df['v1']:
    # Vérifier si le texte est non nul
    if pd.isna(text):
        cleaned_v1_stem.append("")
        continue
    
    # Supprimer les caractères spéciaux et les chiffres
    text = re.sub(r'[^A-Za-z\s]', '', text)
    
    # Convertir le texte en minuscules
    text = text.lower()
    
    # Tokenizer le texte
    words = nltk.word_tokenize(text)
    
    # Éliminer les stopwords et stemmer les mots restants
    stemmed_words = [stemmer.stem(word) for word in words if word not in stop_words]
    
    # Rejoindre les mots en une seule chaîne
    cleaned_text = ' '.join(stemmed_words)
    
    cleaned_v1_stem.append(cleaned_text)

# Nettoyer et stemmer la colonne 'v2'
cleaned_v2_stem = []
for text in df['v2']:
    # Vérifier si le texte est non nul
    if pd.isna(text):
        cleaned_v2_stem.append("")
        continue
    
    # Supprimer les caractères spéciaux et les chiffres
    text = re.sub(r'[^A-Za-z\s]', '', text)
    
    # Convertir le texte en minuscules
    text = text.lower()
    
    # Tokenizer le texte
    words = nltk.word_tokenize(text)
    
    # Éliminer les stopwords et stemmer les mots restants
    stemmed_words = [stemmer.stem(word) for word in words if word not in stop_words]
    
    # Rejoindre les mots en une seule chaîne
    cleaned_text = ' '.join(stemmed_words)
    
    cleaned_v2_stem.append(cleaned_text)

# Ajouter les nouvelles colonnes au dataframe
df['cleaned_v1_stem'] = cleaned_v1_stem
df['cleaned_v2_stem'] = cleaned_v2_stem

# Afficher les résultats
print(df[['v1', 'cleaned_v1_stem', 'v2', 'cleaned_v2_stem']].head())
```
