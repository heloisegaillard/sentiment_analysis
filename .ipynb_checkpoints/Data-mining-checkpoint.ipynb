```python
df=pd.read_csv('spam.csv', encoding='ISO-8859-1')
```

# Nettoyage initial

-Suppression des caractères spéciaux

-Suppression des espaces blancs 

-Suppression des majuscules 

```python
def clean_text(text):
    # Suppression des balises HTML
    text = BeautifulSoup(text, "html.parser").get_text()
    
    # Suppression des caractères spéciaux et des chiffres
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Suppression des espaces blancs supplémentaires
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text.lower() # on met tout le texte en minuscule

# Appliquer le nettoyage initial à chaque texte
df['cleaned_text'] = df['v2'].apply(clean_text)
```

# Suppression des StopWords 

<div style="background-color:lightyellow">
<span>Les **stopwords** varient d'une langue à une autre, mais voici quelques exemples courants en anglais :

    -Articles : "a", "an", "the"
        
    -Pronoms : "I", "you", "he", "she", "it", "they", "we"
        
    -Prépositions : "in", "on", "at", "by", "with", "about", "for"
        
    -Conjonctions : "and", "but", "or", "so", "if", "because"
    </span> </div>

```python
def remove_stopwords(text):
    # Tokenisation du texte
    tokens = word_tokenize(text)
    
    # Liste des stop words en anglais
    stop_words = set(stopwords.words('english'))
    
    # Filtrer les tokens pour supprimer les stop words
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    
    # Rejoindre les tokens filtrés en une chaîne de texte
    filtered_text = ' '.join(filtered_tokens)
    
    return filtered_text

df['text_without_stopwords'] = df['cleaned_text'].apply(remove_stopwords)
```

# Tokenisation

**Tokenisation** :Processus de division d'un texte en unités plus petites appelées tokens, comme des mots ou des symboles, pour faciliter l'analyse linguistique et le traitement automatique du langage naturel (NLP).


-Tokenisation basée sur les espaces

-Tokenisation basée sur les expressions régulières

-Tokenisation basée sur les règles linguistiques

```python
def tokenize_text(text): # tokenisation basée sur les espaces
    tokens = word_tokenize(text)
    return tokens
df['tokens'] = df['text_without_stopwords'].apply(tokenize_text)
```

# Stemmatisation 

**Stemmatisation** : Réduit chaque token à sa racine ou base, comme "running" à "run" ou "cities" à "citi".

```python
def stem_tokens(tokens):
    stemmer = PorterStemmer()
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    return stemmed_tokens

df['stemmed_text'] = df['tokens'].apply(stem_tokens)
```

# Lemmatisation

**Lemmatisation** : Ramène chaque token à sa forme canonique ou lemmes, comme "running" à "run" ou "cities" à "city".

```python
def lemmatize_tokens(tokens):
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return lemmatized_tokens
df['lemmatized_tokens'] = df['tokens'].apply(lemmatize_tokens)
```

## Bigrammes

```python
# Créer une chaîne de tous les tokens lemmatisés séparés par un espace
tokens = df['lemmatized_tokens'].explode().astype(str).tolist()

# Créer une liste de paires de mots (bi-grammes)
bigrams = [(tokens[i], tokens[i + 1]) for i in range(len(tokens) - 1)]

# Compter la fréquence des bi-grammes
bigram_counts = Counter(bigrams)

# Convertir les données en DataFrame
bigram_df = pd.DataFrame(bigram_counts.items(), columns=['Bigram', 'Count'])
bigram_df[['Word1', 'Word2']] = pd.DataFrame(bigram_df['Bigram'].tolist(), index=bigram_df.index)

# Filtrer les bigrammes les plus fréquents
top_n = 20  # Par exemple, les 20 bigrammes les plus fréquents
top_bigrams = bigram_df.nlargest(top_n, 'Count')

# Créer la pivot table
bigram_df_pivot = top_bigrams.pivot(index='Word1', columns='Word2', values='Count').fillna(0).astype(int)

# Créer la heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(bigram_df_pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Heatmap des Bi-grammes Lemmatisés (Top 20) GENERAL')
plt.show()
```
